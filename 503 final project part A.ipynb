```python
df = df[df['DR1_030Z'] == 6]
```

```python
cols = ['DR1ISFAT', 'DR1IMFAT', 'DR1IPFAT', 'DR1IKCAL', 'DR1IPROT', 'DR1ICARB', 'DR1ISUGR', 'DR1IFIBE', 'DR1ITFAT', 'DR1CCMTX']
dietary = df[cols]
dietary.columns = ['Total saturated fatty acids', 'Total monounsaturated fatty acids', 'Total polyunsaturated fatty acids', 'Energy', 'Protein', 'Carbohydrate', 'Sugars', 'Dietary Fiber', 'Total Fat', 'food type']
```

```python
dietary.head()
```

```python
dietary = dietary[dietary['food type'] != 0.0]
dietary = dietary[dietary['food type'] != 7.0]
dietary = dietary[dietary['food type'] != 13.0]
dietary = dietary[dietary['food type'] != 90.0]
dietary['food type'].value_counts()
```

```python
# food type 1: Don't need to heat or cook
dietary['food type'] = dietary['food type'].replace(1.0, 'w/o heat or cook')
dietary['food type'] = dietary['food type'].replace(4.0, 'w/o heat or cook')
dietary['food type'] = dietary['food type'].replace(5.0, 'w/o heat or cook')
dietary['food type'] = dietary['food type'].replace(8.0, 'w/o heat or cook')
dietary['food type'] = dietary['food type'].replace(10.0, 'w/o heat or cook')
dietary['food type'] = dietary['food type'].replace(11.0, 'w/o heat or cook')

dietary['food type'] = dietary['food type'].replace(2.0, 'heat or cook')
dietary['food type'] = dietary['food type'].replace(6.0, 'heat or cook')
dietary['food type'] = dietary['food type'].replace(12.0, 'heat or cook')

dietary['food type'] = dietary['food type'].replace(3.0, 'baking')

dietary['food type'] = dietary['food type'].replace(9.0, 'fried')
dietary['food type'] = dietary['food type'].replace(14.0, 'baking')



```
